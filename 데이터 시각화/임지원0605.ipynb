```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font',family='Malgun Gothic')
%matplotlib inline
```

```python
#1
import csv
f = open('cafe_2year.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
list_data = []

for line in lines:
    list_data.append(line[:])
    
year = []
caffe = []
count = 0
is_first = True

for data in list_data:
    if is_first == True:
        is_first = False
        continue
        
    if count%3 ==0:
        year.append(data[0])
        caffe.append(data[1])
    
    count+=1
    
plt.bar(year, caffe)

font1 = {'size':18,'color':'green'}
plt.title('전국의 월별 카페 수',fontdict=font1)
plt.xlabel('년월')
plt.ylabel('카페 수')

plt.show()

# print(list_data)
f.close()
```

```python
#2
import csv

f = open('doctor_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)

list_data = []

for line in lines:
    list_data.append(line[:])

is_first = True
doctor = {}

for data in list_data:
    if  is_first == True:
        is_first = False
        continue
        
    if data[0] in doctor:
        doctor[data[0]][0] += int(data[2])
        doctor[data[0]][1] += int(data[3])
        doctor[data[0]][2] += int(data[4])
        doctor[data[0]][3] += int(data[5])
    else:
        doctor[data[0]] = [0,0,0,0]
        
locate = ['서울','부산','대구','인천','대전','광주','울산']
print('지역 일반의 인턴 레지던트 전문의')

for key in locate:
    print("%s  %3d  %4d  %4d  %5d"%(key,doctor[key][0],doctor[key][1],doctor[key][2],doctor[key][3]))

f.close()
```

```python
#4
import csv

f = open('doctor2.csv','r',encoding='utf-8')
lines = csv.reader(f)

list_data = []
for line in lines:
    list_data.append(line[:])

area = []
doctor = []

is_first = True
for data in list_data:
    if is_first == True:
        is_first = False
        continue
        
    area.append(data[0])
    doctor.append(int(data[1])+int(data[2])+int(data[3])+int(data[4]))

font1 = {'size':18,'color':'green'}
plt.title('서울 및 광역시 종합병원 전문의 수',fontdict=font1)
plt.scatter(area, doctor)
plt.xlabel('시')
plt.ylabel('의사 수')
plt.show()

f.close()
```

```python
#5
import csv

f = open('doctor2.csv','r',encoding='utf-8')
lines = csv.reader(f)

list_data = []
for line in lines:
    list_data.append(line[:])

area = []
normal = []
intern = []
resident = []
special = []

is_first = True
for data in list_data:
    if is_first == True:
        is_first = False
        continue
    area.append(data[0])
    normal.append(int(data[1]))
    intern.append(int(data[2]))
    resident.append(int(data[3]))
    special.append(int(data[4]))

font1 = {'size':18,'color':'green'}
plt.title('서울 및 광역시 종합병원 전문의 수',fontdict=font1)
plt.plot(area,normal, label='일반의', color ='red', linestyle='-', marker='o', linewidth=2, markersize=12)
plt.plot(area,intern, label='인턴', color ='blue', linestyle='--', marker='x', linewidth=2, markersize=12)
plt.plot(area,resident, label='레지던트', color ='yellow', linestyle=':', marker='s', linewidth=2, markersize=12)
plt.plot(area,special, label='전문의', color ='purple', linestyle='-.', marker='d', linewidth=2, markersize=12)

plt.xlabel('시')
plt.ylabel('의사 수')
plt.show()

f.close()
```
