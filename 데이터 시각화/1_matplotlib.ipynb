```python
#12-4
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
plt.plot(xdata,ydata1,label='국어')
plt.plot(xdata,ydata2,label='영어')
plt.legend(loc='upper center')

plt.title('세명 학생의 국어, 영어 성적')
plt.show()
```

```python
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,78]
ydata4 = [92,88,82]
plt.plot(xdata,ydata1,label='국어', color = 'red', linestyle = '-', marker='o')
plt.plot(xdata,ydata2,label='영어', color = '#00ffff', linestyle = '--', marker = 'x')
plt.plot(xdata,ydata3,label='수학', color = 'magenta', linestyle = '-.', marker = 's')
plt.plot(xdata,ydata4,label='사회', color = '#444444', linestyle = ':', marker = 'd')

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
plt.plot(xdata,ydata1, 'ro--',label='국어', linewidth=3, markersize = 10)
plt.plot(xdata,ydata2,'cx--',label='영어',linewidth=3, markersize = 10)
plt.plot(xdata,ydata3,'ms-.', label = '수학',linewidth=3, markersize = 10)
plt.plot(xdata,ydata4,'kd:', label='사회',linewidth=3, markersize = 10)
plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
xdata = list(range(30))
ydata = []
for x in xdata : 
    y = 2*x
    ydata.append(y)
    
plt.plot(xdata,ydata,color = 'magenta',label = 'y=2x')
plt.xlim(-1,30)
plt.ylim(-2,60)
plt.xticks(list(range(0,30,2)))
plt.yticks(list(range(0,60,5)))

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)

plt.figure(figsize=(10,8))
plt.scatter(X,Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5,1.5), plt.xticks([])
plt.ylim(-1.5,1.5), plt.yticks([])
plt.show()
```

```python
xs = np.linspace(-3, 3, 601)
ys = 1 / (1 + np.exp(-xs))
plt.plot(xs, ys)
plt.ylim(-0.1,1.1)
plt.title('시그모이드 그래프')
plt.axhline(y=0, color = 'k', linewidth=0.5)
plt.axhline(y=1, color = 'k', linewidth=0.5)
plt.axvline(x=0, color = 'k', linewidth=0.5)
plt.show()
```
